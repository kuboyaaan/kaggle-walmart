## ライブラリの読み込み

```python
import os
import numpy as np
import pandas as pd
pd.set_option('max_columns', 100)
import matplotlib.pyplot as plt

# ./walmart/input/配下にコンペデータのディレクトリがある想定
print(os.listdir('./walmart/input/walmart-recruiting-store-sales-forecasting/'))

BASE_DIR = './walmart/input/walmart-recruiting-store-sales-forecasting/'
# kaggle上では以下になる

# BASE_DIR = './kaggle/input/walmart-recruiting-store-sales-forecasting/'
```

```python
train = pd.read_csv(BASE_DIR + 'train.csv.zip')
test = pd.read_csv(BASE_DIR + 'test.csv.zip')
stores = pd.read_csv(BASE_DIR + 'stores.csv')
features = pd.read_csv(BASE_DIR + 'features.csv.zip')
submission = pd.read_csv(BASE_DIR + 'sampleSubmission.csv.zip')
```

```python
# Dept数はStoreごとに異なるかどうか
print('各StoreのDept数最大: ', train.groupby('Store')['Dept'].nunique().max())
print('各StoreのDept数最小: ', train.groupby('Store')['Dept'].nunique().min())

# StoreごとのDept数をヒストグラムで可視化
train.groupby('Store')['Dept'].nunique().hist()
```

```python
# Temperature, Fuel_Price, CPI, Unemployment
# 時系列可視化(Store==1)
import matplotlib
fig, ax = plt.subplots(figsize=(15, 7))
ax2 = ax.twinx()
ax3 = ax.twinx()
ax4 = ax.twinx()
idx = (features['Store'] ==1 ) 
dates = matplotlib.dates.date2num(features.loc[idx, 'Date'])
ax.plot_date(dates, features.loc[idx, 'Temperature'], '-', color='tab:brown', label='Temperature')
ax2.plot_date(dates, features.loc[idx, 'Fuel_Price'], '-', color='tab:cyan', label='Fuel_Price')
ax3.plot_date(dates, features.loc[idx, 'CPI'], '-', color='tab:green', label='CPI')
ax4.plot_date(dates, features.loc[idx, 'Unemployment'], '-', color='tab:red', label='Unemployment')
ax.set_ylabel('Temperature'); ax2.set_ylabel('Fuel Price'); 
ax.legend(loc='upper left'); ax2.legend(loc='upper right'); ax3.legend(loc='lower left'); ax4.legend(loc='lower right');
```
